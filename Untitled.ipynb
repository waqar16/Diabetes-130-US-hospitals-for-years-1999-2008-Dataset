```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split,cross_val_score, StratifiedKFold, KFold
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import np_utils
```

```python
dset = pd.read_csv("../Dataset/diabetic_data.csv")
```

```python
dset.drop(['encounter_id','patient_nbr','weight', 'payer_code', 'medical_specialty'], axis=1, inplace=True)  # Drop the 'weight' column

dset.dropna(subset=['race', 'diag_1', 'diag_2', 'diag_3'], inplace=True)  # Drop rows with null values in the 'race' column
```

```python
dset.isnull().sum()
```

```python
dset.shape
```

```python
# Seperating the feature and target column then i perform encoding

feature_X = dset.iloc[:, 0:44] 
target_y = dset.iloc[:, -1]

target_y = target_y.replace({"NO":0,"<30":1, ">30":0})
```

```python
# Separate categorical and numeric columns in feature_X
categorical_columns = feature_X.select_dtypes(include='object').columns
numeric_columns = feature_X.select_dtypes(include='number').columns

# Label Encoding for Categorical Columns
label_encoder = LabelEncoder()
encoded_categorical_columns = feature_X[categorical_columns].apply(label_encoder.fit_transform)

# Concatenating encoded categorical columns with numeric columns
encoded_feature_X = pd.concat([encoded_categorical_columns, feature_X[numeric_columns]], axis=1)

# Standard Scaling for encoded_feature_X
scaler = StandardScaler()
scaled_encoded_feature_X = scaler.fit_transform(encoded_feature_X)
```

```python
# # Label Encoding for Categorical Columns for the target_y

# label_encoder = LabelEncoder()
# encoded_target_y = label_encoder.fit_transform(target_y)
```

```python
# Perform one-hot encoding
y = np_utils.to_categorical(target_y)
y
```

```python
# Now model is ready to train

no_rows_X, n_columns_X = scaled_encoded_feature_X.shape
no_rows_y, n_columns_y = y.shape

model = keras.Sequential()
model.add(layers.Dense(128, activation='relu', input_shape=(n_columns_X,)))  # input_dim is the number of features
model.add(layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(layers.Dropout(0.2)) 
model.add(layers.Dense(n_columns_y, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# we use earlystop technique because  If the model is trained for too long, it can start to memorize the training examples. Consider using early stopping techniques, where training is halted if the validation loss stops improving, to prevent overfitting.
```

```python
# Define the number of folds for cross-validation
n_folds = 3

# Perform cross-validation
kf = KFold(n_splits=n_folds)
cv_scores = []
history_list = []
for train_index, test_index in kf.split(scaled_encoded_feature_X):
    X_train, X_test = scaled_encoded_feature_X[train_index], scaled_encoded_feature_X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Split the training data into training and validation sets
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)
    
    # Fit the model on the training data with validation set
    history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=40, batch_size=5, verbose=0)
    history_list.append(history)
    
    # Evaluate the model on the test data
    _, accuracy = model.evaluate(X_test, y_test, verbose=0)
    cv_scores.append(accuracy)

    # Plot the training and validation accuracy
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    # Plot the training and validation loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
```

```python
Accuracy = np.mean(cv_scores) * 100
print("Average Accuracy Score: " + str(np.round(Accuracy)) + "%")
```
